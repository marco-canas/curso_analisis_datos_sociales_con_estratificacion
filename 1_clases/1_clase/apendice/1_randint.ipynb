# 3. Respuestas a las preguntas conceptuales



# ¬øQu√© tipo de datos genera?

`np.random.randint()` genera:

* **N√∫meros enteros aleatorios**
* Tipo de dato por defecto:

  ```python
  int64 (o int32 seg√∫n la arquitectura)
  ```


* Devuelve:

  * Un **entero escalar** si `size=None`
  * Un **array de NumPy** si `size` se especifica

Ejemplo para verificar el tipo:


````python
x = np.random.randint(1, 10, size=(2,2,3))
x 
````

````python
x.dtype

````


# ¬øCu√°l es la diferencia entre `low` y `high`?



* `low`:
  - **L√≠mite inferior incluido**
* `high`:
  - **L√≠mite superior excluido**



Formalmente, los valores generados pertenecen al intervalo:

$$
\text{low} \leq X < \text{high}
$$



Ejemplo:


````python
np.random.randint(5, 8, size=20)

````


Valores posibles:

```text
5, 6, 7
```

Nunca aparecer√° el 8.



üìå Caso especial:


````python

np.random.randint(low = 0, high = 10)

````


Aqu√≠:

* `low = 0`
* `high = 10`

---



# ¬øPor qu√© el l√≠mite superior **no se incluye**?

Esta es una decisi√≥n **de dise√±o matem√°tico y computacional**, con varias ventajas:



# 1. Consistencia con Python

Python usa intervalos **cerrado‚Äìabierto**:


````python


list(range(0, 10))     # 0, 1, ..., 9

````


`np.random.randint()` sigue la misma convenci√≥n.

---



# 2. Facilita el conteo de valores posibles



El n√∫mero de enteros posibles es simplemente:

$$
\text{high} - \text{low}
$$

Ejemplo:


````python
np.random.randint(0, 5, size = 20)
# 0, 1, 2, 3, 4
````


Valores posibles ‚Üí 5 valores

```text
0, 1, 2, 3, 4
```
