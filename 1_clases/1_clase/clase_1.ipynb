# Script de Python: creaci√≥n de `encuesta.dta`


````python
import pandas as pd
import numpy as np

# -------------------------------
# 1. Fijar semilla (reproducibilidad)
# -------------------------------
np.random.seed(42)

# -------------------------------
# 2. N√∫mero de observaciones
# -------------------------------
n = 100

# -------------------------------
# 3. Crear variables de encuesta
# -------------------------------

df = pd.DataFrame({

    # Identificador
    "id": range(1, n + 1),

    # Variables sociodemogr√°ficas
    "edad": np.random.randint(18, 65, size=n),

    # 1 = Mujer, 2 = Hombre
    "genero": np.random.choice([1, 2], size=n, p=[0.52, 0.48]),

    # Nivel educativo (ordinal)
    # 1 = Primaria
    # 2 = Secundaria
    # 3 = T√©cnica
    # 4 = Universitaria
    "educacion": np.random.choice([1, 2, 3, 4], size=n, p=[0.25, 0.35, 0.2, 0.2]),

    # Clase social (esquema simplificado)
    # 1 = Clase baja
    # 2 = Clase media
    # 3 = Clase alta
    "clase_social": np.random.choice([1, 2, 3], size=n, p=[0.4, 0.45, 0.15]),

    # Ingreso mensual (en pesos)
    "ingreso": np.random.normal(loc=2_500_000, scale=800_000, size=n).astype(int),

    # Condici√≥n de ocupaci√≥n
    # 1 = Ocupado, 0 = No ocupado
    "ocupado": np.random.choice([0, 1], size=n, p=[0.25, 0.75])

})
df.head() 
````

````python
# -------------------------------
# 4. Asegurar ingresos positivos
# -------------------------------
df["ingreso"] = df["ingreso"].clip(lower=500000) # acotar inferiormente los ingresos 
df.head() 
````

````python
# -------------------------------
# 5. Guardar archivo en formato Stata
# -------------------------------
df.to_stata("encuesta.dta", write_index=False)

# -------------------------------
# 6. Verificaci√≥n r√°pida
# -------------------------------
df.head()

````

# Interpretaci√≥n de resultados anteriores  

Te doy una **interpretaci√≥n pedag√≥gica, sociol√≥gicamente orientada y par√°metro por par√°metro** del **resumen (`summary`) de una regresi√≥n lineal** estimada con `statsmodels`, exactamente en el nivel que necesitan tus estudiantes para **entender qu√© hace Stata y c√≥mo Python replica lo mismo**.

Partimos del modelo:


````python


import statsmodels.formula.api as smf

modelo = smf.ols("ingreso ~ educacion", data=df).fit()
modelo.summary()

````


Este modelo estima una **regresi√≥n lineal simple** donde:

* **Variable dependiente (Y):** `ingreso`
* **Variable independiente (X):** `educacion`

---



# 1. ¬øQu√© est√° estimando este modelo?

El modelo estad√≠stico es:

$$
\text{ingreso}_i = \beta_0 + \beta_1 \cdot \text{educacion}_i + \varepsilon_i
$$

Interpretaci√≥n sociol√≥gica:

> Estamos evaluando **c√≥mo cambia el ingreso promedio** cuando aumenta el nivel educativo, asumiendo una relaci√≥n lineal y promedio en la poblaci√≥n observada.

---



# 2. Estructura general del `summary()`

El resumen se divide en **cuatro grandes bloques**:

1. Informaci√≥n general del modelo
2. Calidad global del ajuste
3. Tabla de coeficientes
4. Diagn√≥sticos y supuestos

Voy bloque por bloque.

---



## 3. Bloque 1: Informaci√≥n general del modelo

Ejemplo t√≠pico:

```
Dep. Variable:                ingreso
Model:                            OLS
Method:                 Least Squares
Date:                ...
Time:                ...
No. Observations:                 100
Df Residuals:                      98
Df Model:                           1
```



### Interpretaci√≥n

* **Dep. Variable:**
  Variable dependiente ‚Üí `ingreso`

* **Model:**
  OLS (*Ordinary Least Squares*) ‚Üí regresi√≥n lineal cl√°sica

* **Method:**
  M√≠nimos Cuadrados Ordinarios

* **No. Observations:**
  N√∫mero de casos usados en el modelo

* **Df Model:**
  N√∫mero de variables independientes (1: `educacion`)

* **Df Residuals:**
  Grados de libertad de los residuos
  $$
  n - k - 1
  $$

---



## 4. Bloque 2: Calidad global del modelo

Ejemplo:

```
R-squared:                       0.35
Adj. R-squared:                  0.34
F-statistic:                    52.3
Prob (F-statistic):           0.0000
```



### Par√°metros clave

### üîπ R-squared (R¬≤)

* Proporci√≥n de la variabilidad del ingreso explicada por educaci√≥n.
* Ejemplo: `R¬≤ = 0.35`
  ‚Üí El **35 % de la desigualdad del ingreso** se asocia al nivel educativo.

‚ö†Ô∏è No implica causalidad.

---



### üîπ Adjusted R-squared

* Versi√≥n corregida por n√∫mero de variables.
* √ötil en modelos m√∫ltiples.
* Aqu√≠ ser√° muy similar al R¬≤.

---



### üîπ F-statistic

* Prueba global del modelo:

  > ¬øEl modelo explica m√°s que un modelo sin predictores?

---



### üîπ Prob (F-statistic)

* Valor p del test F.
* Si `p < 0.05`:
  ‚Üí El modelo es **estad√≠sticamente significativo** en conjunto.

---



## 5. Bloque 3: Tabla de coeficientes (el coraz√≥n del an√°lisis)

Ejemplo t√≠pico:

```
==============================================================================
                 coef    std err          t      P>|t|      [0.025      0.975]
------------------------------------------------------------------------------
Intercept   800000.0  250000.0      3.20      0.002    300000.0   1300000.0
educacion   450000.0   62000.0      7.23      0.000    325000.0    575000.0
==============================================================================
```

Veamos **columna por columna**.

---



## Intercept (Œ≤‚ÇÄ)

* **coef:** ingreso esperado cuando `educacion = 0`
* En encuestas reales:

  * Puede no tener sentido sustantivo
  * Es un punto de referencia matem√°tico

* Sociol√≥gicamente:

> Es el ingreso promedio del grupo con nivel educativo m√≠nimo (o fuera de escala).

---



## educacion (Œ≤‚ÇÅ)

* **coef:** cambio promedio en ingreso por cada aumento de una unidad en educaci√≥n.
* Ejemplo:

  > Cada nivel educativo adicional se asocia con **$450.000 pesos m√°s** de ingreso mensual.

---



## std err (error est√°ndar)

* Mide la **incertidumbre** del coeficiente.
* Menor error ‚Üí estimaci√≥n m√°s precisa.

---



## t (estad√≠stico t)

$$
t = \frac{\text{coef}}{\text{std err}}
$$



* Mide cu√°ntos errores est√°ndar se aleja de cero.

---



## P>|t| (valor p)

* Probabilidad de observar ese coeficiente si el verdadero efecto fuera cero.
* Regla pr√°ctica:

  * `p < 0.05` ‚Üí evidencia estad√≠stica de asociaci√≥n

---



## [0.025, 0.975] (intervalo de confianza 95 %)

* Rango plausible del verdadero coeficiente.
* Si **no incluye el 0**, el efecto es significativo.

---



## 6. Bloque 4: Diagn√≥sticos del modelo

Ejemplo:

```
Omnibus:                        1.85
Prob(Omnibus):                  0.40
Durbin-Watson:                  2.01
Jarque-Bera (JB):               1.70
Prob(JB):                       0.42
Skew:                           0.15
Kurtosis:                       2.90
```



# Interpretaci√≥n (nivel introductorio)

* **Durbin-Watson (~2):**

  * No hay autocorrelaci√≥n fuerte en residuos.



* **Omnibus / Jarque-Bera:**

  * Pruebas de normalidad de residuos.
  * Si p > 0.05 ‚Üí no hay evidencia fuerte de no-normalidad.

* Para cursos introductorios:

> Se reportan, pero **no se profundiza demasiado**.

## 2. Script completo en Python con matplotlib


````python

import numpy as np
import matplotlib.pyplot as plt

# Variable independiente
X = df["educacion"]
# Variable dependiente
y = df["ingreso"]

# Valores predichos por el modelo
y_pred = modelo.predict(df)

# Crear la figura
plt.figure(figsize=(10, 6))

# Gr√°fico de dispersi√≥n: datos reales
plt.scatter(X, y)

# Recta de regresi√≥n
# Ordenamos para que la l√≠nea se vea continua
orden = np.argsort(X)
plt.plot(X.iloc[orden], y_pred.iloc[orden])

# Etiquetas y t√≠tulo
plt.xlabel("Nivel educativo")
plt.ylabel("Ingreso")
plt.title("Ingreso vs Educaci√≥n con recta de regresi√≥n OLS")
plt.savefig(r"C:\Users\marco\Downloads\ingreso_educacion_regresion.png")
# Mostrar
plt.show()

````
