# Script de Python: creación de `encuesta.dta`


````python
import pandas as pd
import numpy as np
````

````python

# -------------------------------
# 1. Fijar semilla (reproducibilidad)
# -------------------------------
np.random.seed(42)
````

````python

# -------------------------------
# 2. Número de observaciones
# -------------------------------
n = 100
````

````python

# -------------------------------
# 3. Crear variables de encuesta
# -------------------------------

df = pd.DataFrame({

    # Identificador
    "id": range(1, n + 1),

    # Variables sociodemográficas
    "edad": np.random.randint(18, 65, size=n),

    # 1 = Mujer, 2 = Hombre
    "genero": np.random.choice([1, 2], size=n, p=[0.52, 0.48]),

    # Nivel educativo (ordinal)
    # 1 = Primaria
    # 2 = Secundaria
    # 3 = Técnica
    # 4 = Universitaria
    "educacion": np.random.choice([1, 2, 3, 4], size=n, p=[0.25, 0.35, 0.2, 0.2]),

    # Clase social (esquema simplificado)
    # 1 = Clase baja
    # 2 = Clase media
    # 3 = Clase alta
    "clase_social": np.random.choice([1, 2, 3], size=n, p=[0.4, 0.45, 0.15]),

    # Ingreso mensual (en pesos)
    "ingreso": np.random.normal(loc=2_500_000, scale=800_000, size=n).astype(int),

    # Condición de ocupación
    # 1 = Ocupado, 0 = No ocupado
    "ocupado": np.random.choice([0, 1], size=n, p=[0.25, 0.75])

})
df.head() 
````

````python
df.to_stata("datos_encuesta.dta", write_index=False)
````

````python

import pandas as pd 
df = pd.read_stata("encuesta.dta")
df.head()
````

# Interpretación de resultados anteriores  

Te doy una **interpretación pedagógica, sociológicamente orientada y parámetro por parámetro** del **resumen (`summary`) de una regresión lineal** estimada con `statsmodels`, exactamente en el nivel que necesitan tus estudiantes para **entender qué hace Stata y cómo Python replica lo mismo**.

Partimos del modelo:


````python


import statsmodels.formula.api as smf

modelo = smf.ols("ingreso ~ educacion", data=df).fit()
modelo.summary()

````


Este modelo estima una **regresión lineal simple** donde:

* **Variable dependiente (Y):** `ingreso`
* **Variable independiente (X):** `educacion`

---



# 1. ¿Qué está estimando este modelo?

El modelo estadístico es:

$$
\text{ingreso}_i = \beta_0 + \beta_1 \cdot \text{educacion}_i + \varepsilon_i
$$

Interpretación sociológica:

> Estamos evaluando **cómo cambia el ingreso promedio** cuando aumenta el nivel educativo, asumiendo una relación lineal y promedio en la población observada.

---



# 2. Estructura general del `summary()`

El resumen se divide en **cuatro grandes bloques**:

1. Información general del modelo
2. Calidad global del ajuste
3. Tabla de coeficientes
4. Diagnósticos y supuestos

Voy bloque por bloque.

---



## 3. Bloque 1: Información general del modelo

Ejemplo típico:

```
Dep. Variable:                ingreso
Model:                            OLS
Method:                 Least Squares
Date:                ...
Time:                ...
No. Observations:                 100
Df Residuals:                      98
Df Model:                           1
```



### Interpretación

* **Dep. Variable:**
  Variable dependiente → `ingreso`

* **Model:**
  OLS (*Ordinary Least Squares*) → regresión lineal clásica

* **Method:**
  Mínimos Cuadrados Ordinarios

* **No. Observations:**
  Número de casos usados en el modelo

* **Df Model:**
  Número de variables independientes (1: `educacion`)

* **Df Residuals:**
  Grados de libertad de los residuos
  $$
  n - k - 1
  $$

---



## 4. Bloque 2: Calidad global del modelo

Ejemplo:

```
R-squared:                       0.35
Adj. R-squared:                  0.34
F-statistic:                    52.3
Prob (F-statistic):           0.0000
```



# Parámetros clave



# R-squared (R²)



* Proporción de la variabilidad del ingreso explicada por educación.
* Ejemplo: `R² = 0.35`
  → El **35 % de la desigualdad del ingreso** se asocia al nivel educativo.

* No implica causalidad.





## Adjusted R-squared



* Versión corregida por número de variables.
* Útil en modelos múltiples.
* Aquí será muy similar al R².

---



## F-statistic

* Prueba global del modelo:

  > ¿El modelo explica más que un modelo sin predictores?

---



## Prob (F-statistic)

* Valor p del test F.
* Si `p < 0.05`:
  → El modelo es **estadísticamente significativo** en conjunto.

---



## 5. Bloque 3: Tabla de coeficientes (el corazón del análisis)

Ejemplo típico:

```
==============================================================================
                 coef    std err          t      P>|t|      [0.025      0.975]
------------------------------------------------------------------------------
Intercept   800000.0  250000.0      3.20      0.002    300000.0   1300000.0
educacion   450000.0   62000.0      7.23      0.000    325000.0    575000.0
==============================================================================
```

Veamos **columna por columna**.

---



## Intercept (β₀)

* **coef:** ingreso esperado cuando `educacion = 0`
* En encuestas reales:

  * Puede no tener sentido sustantivo
  * Es un punto de referencia matemático

* Sociológicamente:

> Es el ingreso promedio del grupo con nivel educativo mínimo (o fuera de escala).

---



## educacion (β₁)

* **coef:** cambio promedio en ingreso por cada aumento de una unidad en educación.
* Ejemplo:

  > Cada nivel educativo adicional se asocia con **$450.000 pesos más** de ingreso mensual.

---



## std err (error estándar)

* Mide la **incertidumbre** del coeficiente.
* Menor error → estimación más precisa.

---



## t (estadístico t)

$$
t = \frac{\text{coef}}{\text{std err}}
$$



* Mide cuántos errores estándar se aleja de cero.

---



## P>|t| (valor p)

* Probabilidad de observar ese coeficiente si el verdadero efecto fuera cero.
* Regla práctica:

  * `p < 0.05` → evidencia estadística de asociación

---



## [0.025, 0.975] (intervalo de confianza 95 %)

* Rango plausible del verdadero coeficiente.
* Si **no incluye el 0**, el efecto es significativo.

---



## 6. Bloque 4: Diagnósticos del modelo

Ejemplo:

```
Omnibus:                        1.85
Prob(Omnibus):                  0.40
Durbin-Watson:                  2.01
Jarque-Bera (JB):               1.70
Prob(JB):                       0.42
Skew:                           0.15
Kurtosis:                       2.90
```



# Interpretación (nivel introductorio)

* **Durbin-Watson (~2):**

  * No hay autocorrelación fuerte en residuos.



* **Omnibus / Jarque-Bera:**

  * Pruebas de normalidad de residuos.
  * Si p > 0.05 → no hay evidencia fuerte de no-normalidad.

* Para cursos introductorios:

> Se reportan, pero **no se profundiza demasiado**.

Esta es una **visualización estándar, clara y didácticamente equivalente a la que verían en Stata**, usando **solo matplotlib**, donde aparecen:

1. **Los datos observados** (`educacion` vs `ingreso`)
2. **La recta estimada por el modelo OLS**

Asumo que ya tienes:

* Un `DataFrame df`
* El modelo ajustado:


````python
import statsmodels.formula.api as smf
modelo = smf.ols("ingreso ~ educacion", data=df).fit()

````


---

## 1. Idea estadística de la visualización

Lo que vamos a graficar es:

$$
\hat{y} = \hat{\beta}_0 + \hat{\beta}_1 x
$$

* **Puntos** → datos reales de la encuesta
* **Línea** → ingreso promedio estimado por el modelo



Esto responde visualmente a la pregunta:

> ¿La educación está asociada con mayores ingresos y cómo de fuerte es esa relación?

---

## 2. Script completo en Python con matplotlib


````python

import numpy as np
import matplotlib.pyplot as plt

# Variable independiente
X = df["educacion"]
# Variable dependiente
y = df["ingreso"]

# Valores predichos por el modelo
y_pred = modelo.predict(df)

# Crear la figura
plt.figure(figsize=(10, 6))

# Gráfico de dispersión: datos reales
plt.scatter(X, y)

# Recta de regresión
# Ordenamos para que la línea se vea continua
orden = np.argsort(X)
plt.plot(X.iloc[orden], y_pred.iloc[orden])

# Etiquetas y título
plt.xlabel("Nivel educativo")
plt.ylabel("Ingreso")
plt.title("Ingreso vs Educación con recta de regresión OLS")
plt.savefig(r"C:\Users\marco\Downloads\ingreso_educacion_regresion.png")
# Mostrar
plt.show()

````


## 3. Cómo explicar este gráfico en clase



## Puntos

Cada punto representa:

* Una persona (u hogar)
* Con un nivel educativo específico
* Y un ingreso observado



La dispersión vertical muestra:

> Que personas con el mismo nivel educativo pueden tener ingresos distintos
> (clase social, género, informalidad, experiencia, etc.).

---



## Línea de regresión

La línea representa:

> El **ingreso promedio esperado** para cada nivel educativo según el modelo.

* Pendiente positiva → asociación positiva
* Pendiente pronunciada → efecto fuerte
* Pendiente suave → efecto débil

---



# 4. Lectura sociológica rápida (ejemplo)

> Aunque se observa una tendencia positiva entre educación e ingreso, la alta dispersión de los puntos sugiere que la educación no es el único determinante del ingreso. Esto justifica la inclusión de variables adicionales como género, clase social o sector laboral en modelos posteriores.

# Práctica de codificación sobre cómo crear un archivo .dta de Stata con Python 



````python
from numpy.random import seed, choice, randint, normal 
import pandas as pd 
````

````python
n = 10 
df = pd.DataFrame({
    "id": range(1,n+1),
    "edad": randint(18, 65, size=n),
    "genero": choice([1, 2], size=n, p=[0.40, 0.60]),
    "educacion": choice([1, 2, 3, 4], size=n, p=[0.25, 0.35, 0.2, 0.2]),
    "clase_social": choice([1, 2, 3], size=n, p=[0.40, 0.40, 0.20]),
    "ingreso": normal(loc=2_500_000, scale=800_000, size=n).astype(int),
    "ocupado": choice([0, 1], size=n, p=[0.25, 0.75])
})
````

````python
df.head() 
df.index = df.id
df = df.drop(columns=["id"])
````

````python
df.to_stata("datos_prueba.dta", write_index=True)
````
