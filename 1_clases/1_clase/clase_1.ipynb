### 4.6 Regresi√≥n lineal simple

**Stata**

```stata
reg ingreso educacion
```

**Python (statsmodels)**

```python
import statsmodels.formula.api as smf

modelo = smf.ols("ingreso ~ educacion", data=df).fit()
modelo.summary()
```

üëâ `statsmodels` replica casi exactamente la salida de Stata.

---

### 4.7 Regresi√≥n lineal m√∫ltiple

**Stata**

```stata
reg ingreso educacion genero edad
```

**Python**

```python
modelo = smf.ols(
    "ingreso ~ educacion + genero + edad",
    data=df
).fit()

modelo.summary()
```

---

### 4.8 Regresi√≥n log√≠stica

**Stata**

```stata
logit ocupado educacion genero edad
```

**Python**

```python
modelo_logit = smf.logit(
    "ocupado ~ educacion + genero + edad",
    data=df
).fit()

modelo_logit.summary()
```
