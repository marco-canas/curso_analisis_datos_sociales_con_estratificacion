# Lectura experimental e interactiva del m贸dulo `smf` de `statsmodels`

## Objetivo

Explorar de forma **experimental e interactiva** el m贸dulo `statsmodels.formula.api (smf)` y, en particular, el m茅todo `smf.ols()`, para **entrenar modelos de regresi贸n lineal** aplicados al **an谩lisis de datos sociales y procesos de estratificaci贸n**.

---



## 1. Contexto conceptual

En ciencias sociales es com煤n modelar relaciones como:

* Ingreso vs. educaci贸n
* Satisfacci贸n con la vida vs. variables socioecon贸micas
* Desigualdad vs. capital humano



El enfoque **OLS (Ordinary Least Squares)** permite estimar relaciones lineales interpretables, esenciales para la inferencia social.



`statsmodels` es especialmente adecuado porque:

* Ofrece **estad铆sticos cl谩sicos** (p-valores, intervalos de confianza)

---



## 2. Preparaci贸n del entorno



```python
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
```

---



## 3. Construcci贸n de un dataset social simulado

Simulamos un conjunto de datos t铆pico de encuesta social.

```python
np.random.seed(42)

n = 200

df = pd.DataFrame({
    "edad": np.random.randint(18, 65, n),
    "educacion": np.random.randint(0, 20, n),  # a帽os de educaci贸n
    "genero": np.random.choice(["Hombre", "Mujer"], n),
})

# Ingreso mensual (modelo latente)
df["ingreso"] = (
    500
    + 80 * df["educacion"]
    + 5 * df["edad"]
    + np.where(df["genero"] == "Hombre", 150, 0)
    + np.random.normal(0, 200, n)
)

df.head()
```

---



## 4. Primer modelo con `smf.ols()`

### Modelo: ingreso ~ educaci贸n

```python
modelo1 = smf.ols(
    formula="ingreso ~ educacion",
    data=df
).fit()

modelo1.summary()
```

### Interpretaci贸n social

* **Intercepto**: ingreso base sin educaci贸n
* **Coeficiente educaci贸n**: retorno econ贸mico por a帽o adicional de estudio

---



## 5. Modelo multivariable: edad y educaci贸n

```python
modelo2 = smf.ols(
    formula="ingreso ~ educacion + edad",
    data=df
).fit()

modelo2.summary()
```

**Discusi贸n**:

* 驴La educaci贸n sigue siendo significativa al controlar por edad?
* 驴Cu谩l variable tiene mayor impacto?

---



## 6. Inclusi贸n de variables categ贸ricas (estratificaci贸n)

`smf` maneja autom谩ticamente variables categ贸ricas.

```python
modelo3 = smf.ols(
    formula="ingreso ~ educacion + edad + genero",
    data=df
).fit()

modelo3.summary()
```

### Lectura sociol贸gica

* El coeficiente de `genero[T.Mujer]` mide **brecha salarial**
* Se interpreta respecto a la categor铆a base (Hombre)

---



## 7. Visualizaci贸n del ajuste

```python
plt.figure(figsize=(8, 5))
plt.scatter(df["educacion"], df["ingreso"], alpha=0.5)
plt.plot(
    df["educacion"],
    modelo1.predict(df),
)
plt.xlabel("A帽os de educaci贸n")
plt.ylabel("Ingreso mensual")
plt.title("Ingreso vs Educaci贸n (OLS)")
plt.show()
```

---



## 8. Predicci贸n y escenarios sociales

```python
escenarios = pd.DataFrame({
    "educacion": [5, 10, 15],
    "edad": [30, 30, 30],
    "genero": ["Mujer", "Mujer", "Mujer"],
})

modelo3.predict(escenarios)
```

**Pregunta gu铆a**:

> 驴C贸mo cambia el ingreso esperado al aumentar la educaci贸n manteniendo constantes las dem谩s variables?

## 9. Diagn贸stico b谩sico del modelo

```python
modelo3.resid.plot(kind="hist", bins=20)
plt.title("Distribuci贸n de residuos")
plt.show()
```
