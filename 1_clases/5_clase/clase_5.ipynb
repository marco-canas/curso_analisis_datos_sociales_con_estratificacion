## ğŸ“ 4. Ajuste del modelo de regresiÃ³n lineal simple

```python
modelo = smf.ols("ingreso ~ educacion", data=df).fit()
modelo.summary()
```

### b) Pendiente (Î²â‚)

## ğŸ“ˆ 6. VisualizaciÃ³n del modelo ajustado

```python
df["ingreso_pred"] = modelo.predict(df)

plt.figure(figsize=(8, 6))
plt.scatter(df["educacion"], df["ingreso"])
plt.plot(df["educacion"], df["ingreso_pred"])
plt.xlabel("Nivel educativo")
plt.ylabel("Ingreso")
plt.title("RegresiÃ³n lineal simple: ingreso ~ educaciÃ³n")
plt.show()
```
